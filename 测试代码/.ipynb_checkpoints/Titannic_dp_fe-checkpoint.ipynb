```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

path = r"../测试数据/train.csv"
data = pd.read_csv(path)
data.head()
```

```python
data.info()
```

# 属性字段说明
* PassengerId：乘客ID
* Survived：是否幸存，幸存为1，遇难为0
* Pclass：客舱等级，1为最高，3为最低
* Name：乘客姓名
* Sex：乘客性别                               &
* Age：乘客年龄
* SibSp：兄弟姐妹的个数
* Parch：父母孩子的个数
* Ticket：乘客票号
* Fare：票价
* Cabin：船舱编号
* Embarked：登船港口

# 绘制存活比例

```python
plt.axis("equal")
groupby_survived = data.groupby('Survived')['Survived'].value_counts()
groupby_survived.plot.pie(autopct = '%1.2f%%')
plt.show()
```

# 缺失值处理

```python
# 统计年龄平均值
Age_values = {'Age': data['Age'].mean()}
# 使用年龄平均值对缺失值进行填充
data = data.fillna(value = Age_values)

# 统计Embarked属性的众数
Embarked_values = {'Embarked': data['Embarked'].mode()[0]}
# 使用众数填充缺失值
data = data.fillna(value = Embarked_values)

# cabin缺失值较多，将缺失表示称一种属性
Cabin_values = {'Cabin': 'U0'}
# 对缺失值填充
data = data.fillna(value = Cabin_values)

data.info()
```

# 对性别变量进行分析及处理

```python
groupby_sex_survived_count = data.groupby(['Sex','Survived'])['Sex'].count()
groupby_sex_survived_count.unstack('Sex').plot.bar(stacked = True,title = 'Distribution of survived by sex')
plt.show()
groupby_sex_survived_rate = data[['Sex','Survived']].groupby(['Sex']).mean()
print(groupby_sex_survived_rate)
```

# onehot处理Sex项

```python
sex_dummies = pd.get_dummies(data['Sex'])
data = data.join(sex_dummies)
data = data.drop(['Sex'], axis = 1)
data.head()
```

# 对姓名变量进行分析处理

```python
import numpy as np

def count_size(a):
    return len(a)

# 这里注意新的属性名字是Name与原属性重名，下面要先删除原属性再执行join操作
name_dummies = data['Name'].str.split(" ").apply(count_size)
data = data.drop(['Name'], axis = 1)

data = data.join(name_dummies)
data.info()
```

```python
data.rename(columns={"Name":"NameLength"},inplace=True)
data.head()
```

# 对船舱等级变量进行分析与处理

```python
groupby_pclass = data["Pclass"].value_counts()
groupby_pclass.plot.bar()
plt.show()
```

```python
groupby_pclass_survived = data.groupby(["Pclass","Survived"])["Pclass"].count()
groupby_pclass_survived.unstack("Survived").plot.bar(stacked = True)
```

```python
groupby_pclass_survived_ratio = data[["Pclass","Survived"]].groupby("Pclass").mean()
print(groupby_pclass_survived_ratio)
```

```python
pclass_dummies = pd.get_dummies(data["Pclass"])
data = data.join(pclass_dummies)
data.rename(columns={1:"pclass_1",2:"pclass_2",3:"pclass_3"},inplace=True)
data.head()
```

```python
data = data.drop(['Pclass'], axis = 1)
data.info()
```

# 对年龄变量进行分析与处理

```python
# 年龄分布
data['Age'].hist(bins=20)
plt.xlabel("Age")
plt.ylabel("Num")
plt.show()
```

```python
data["Age"].describe()
```

```python
bins = [0,12,18,65,100]
data['Age_group'] = pd.cut(data["Age"],bins)
groupby_age_survived_rate = data.groupby('Age_group')["Survived"].mean()
groupby_age_survived_rate.plot.bar()
plt.show()
```

```python
data['Age_group_id'] = pd.factorize(data['Age_group'])[0]
Age_dummies  = pd.get_dummies(data['Age_group_id'])
data = data.join(Age_dummies)
data = data.drop(['Age_group_id','Age_group'], axis=1)

data.rename(columns={0:'Age_bin_0'},inplace=True)
data.rename(columns={1:'Age_bin_1'},inplace=True)
data.rename(columns={2:'Age_bin_2'},inplace=True)
data.rename(columns={3:'Age_bin_3'},inplace=True)
data.head()
```

```python
from sklearn import preprocessing

# 对年龄进行归一化处理
scaler = preprocessing.StandardScaler()
data['Age_scaled'] = scaler.fit_transform(data['Age'].values.reshape(-1, 1))
data.drop(columns=['Age'],axis=1)
data.head()
```

# 对亲友数量进行分析与处理

```python
# 将数据分为有兄弟姐妹的和没有兄弟姐妹的两组
sibsp_df = data[data['SibSp'] != 0]
no_sibsp_df = data[data['SibSp'] == 0]


# 有无兄弟姐妹和生存的关系
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.axis('equal')
groupby_sibsp = sibsp_df['Survived'].value_counts()
groupby_sibsp.plot.pie(labels=['No Survived', 'Survived'], autopct = '%1.1f%%')

plt.subplot(122)
plt.axis('equal')
groupby_no_sibsp = no_sibsp_df['Survived'].value_counts()
groupby_no_sibsp.plot.pie(labels=['No Survived', 'Survived'], autopct = '%1.1f%%')
plt.show()
```

```python
# 兄弟姐妹数量与生存的关系

groupby_sibsp_survived_count = data.groupby(['SibSp','Survived'])['SibSp'].count()
groupby_sibsp_survived_count.unstack('Survived').plot.bar(stacked = True,title = 'Distribution of Survived by Sibsp')
plt.show()

# 统计兄弟姐妹数量下的生存比例
groupby_sibsp_survived_rate = data[['SibSp','Survived']].groupby(['SibSp']).mean()
print(groupby_sibsp_survived_rate)
```

```python
# 有无父母子女与生存的关系

groupby_sibsp_survived_count = data.groupby(['Parch','Survived'])['Parch'].count()
groupby_sibsp_survived_count.unstack('Survived').plot.bar(stacked = True,title = 'Distribution of Survived by Parch')
plt.show()

# 统计父母子女数量与的生存比例
groupby_sibsp_survived_rate = data[['Parch','Survived']].groupby(['Parch']).mean()
print(groupby_sibsp_survived_rate)
```

```python
# 两个属性可以作为一个整体，反映家庭规模的大小

def family_size_category(family_size):
    if family_size <= 1:
        return 'Single'
    elif family_size <= 4:
        return 'Small_Family'
    else:
        return 'Large_Family'

data['Family_Size'] = data['Parch'] + data['SibSp'] + 1
data['Family_Size_Category'] = data['Family_Size'].map(family_size_category)
data.head(3)
```

```python
from sklearn import preprocessing

# 对亲友数进行归一化处理
scaler = preprocessing.StandardScaler()
data['Family_size_scaled'] = scaler.fit_transform(data['Family_Size'].values.reshape(-1, 1))
data.drop(columns=['Family_Size'],axis = 1)
data.head()
```

```python
# onehot处理家庭规模项

family_dummies  = pd.get_dummies(data['Family_Size_Category'])
data = data.join(family_dummies)
data = data.drop(['Family_Size_Category','Family_Size'], axis=1)
data.head(3)
```

# 对票价进行分析与处理

```python
# 票价的分布
plt.figure(figsize=(10,5))
data['Fare'].hist(bins = 70)
plt.show()

print(data['Fare'].describe())
```

```python
# 将数据分为生存的和死亡的两组
survived_data = data[data['Survived'] == 1]
no_survived_data = data[data['Survived'] == 0]

# 统计两组平均票价
groupby_fare = survived_data['Fare'].mean()
no_groupby_fare = no_survived_data['Fare'].mean()

print("幸存者的票价：\n",groupby_fare)
print("遇难者的票价：\n",no_groupby_fare)
```

```python
# 一样的思路对票价进行分桶
data['Fare_bin'] = pd.qcut(data['Fare'], 5)
data['Fare_bin_id'] = pd.factorize(data['Fare_bin'])[0]
Fare_dummies  = pd.get_dummies(data['Fare_bin_id'])
data = data.join(Fare_dummies)
data = data.drop(['Fare_bin_id','Fare_bin'], axis=1)

data.rename(columns={0:'Fare_bin_0'},inplace=True)
data.rename(columns={1:'Fare_bin_1'},inplace=True)
data.rename(columns={2:'Fare_bin_2'},inplace=True)
data.rename(columns={3:'Fare_bin_3'},inplace=True)
data.rename(columns={4:'Fare_bin_4'},inplace=True)
```

```python
from sklearn import preprocessing

# 对票价进行归一化处理
scaler = preprocessing.StandardScaler()
data['Fare_scaled'] = scaler.fit_transform(data['Fare'].values.reshape(-1, 1))
data.drop(columns=['Fare'],axis = 1)
data.head()
```

# 对船舱编号进行分析和处理

```python
# 船舱类型与生存的关系
import re
data['CabinLetter'] = data['Cabin'].map(lambda x: re.compile("([a-zA-Z]+)").search(x).group())

# 不同船舱类型下的生存比例
groupby_cabin_survived_rate = data[['CabinLetter','Survived']].groupby(['CabinLetter']).mean()
groupby_cabin_survived_rate.plot.bar()
plt.show()
```

```python
# 统计有船舱号和无船舱号的数据
no_set_data = data[data["Cabin"]== 'U0']
set_data = data[data["Cabin"]!='U0']

# 统计两组平均幸存率
groupby_no_set_survived = no_set_data['Survived'].mean()
groupby_set_survived = set_data['Survived'].mean()
data = data.drop(columns=['CabinLetter'],axis = 1)

print("无船舱号的平均幸存率：\n",groupby_no_set_survived)
print("有船舱号的平均幸存率：\n",groupby_set_survived)
```

```python
# 有船舱号的分为一类，无船舱号的分为一类
data['Cabin_YN'] = data['Cabin'].apply(lambda x: 0 if x == 'U0' else 1)
data.head(3)
```

```python
# 上船港口和存活与否的关系

groupby_embarked_survived_count = data.groupby(['Embarked','Survived'])['Embarked'].count()
groupby_embarked_survived_count.unstack('Survived').plot.bar(stacked = True,title = 'Distribution of Survived by Embarked')
plt.show()

# 统计不同上船港口的生存比例
groupby_embarked_survived_rate = data[['Embarked','Survived']].groupby(['Embarked']).mean()
print(groupby_embarked_survived_rate)
```

```python
# onehot处理Embarked项
embark_dummies  = pd.get_dummies(data['Embarked'])
data = data.join(embark_dummies)
data = data.drop(['Embarked'], axis=1)

data.rename(columns={'C':'Embark_C'},inplace=True)
data.rename(columns={'Q':'Embark_Q'},inplace=True)
data.rename(columns={'S':'Embark_S'},inplace=True)

data.head(3)
```

```python
data_ = data[['Fare', 'female', 'male', 'NameLength', 'pclass_1', 'pclass_2',
       'pclass_3', 'Age_bin_0', 'Age_bin_1', 'Age_bin_2', 'Age_bin_3',
       'Age_scaled', 'Family_size_scaled', 'Large_Family', 'Single',
       'Small_Family', 'Fare_bin_0', 'Fare_bin_1', 'Fare_bin_2', 'Fare_bin_3',
       'Fare_bin_4', 'Fare_scaled', 'Cabin_YN', 'Embark_C', 'Embark_Q',
       'Embark_S']]
label = data[["Survived"]]
data_ = data
data_.head(5)
```

```python
from sklearn.model_selection import train_test_split

label = np.squeeze(label)
X_train, X_test, y_train, y_test = train_test_split(data_, label, random_state=1,shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("准确率：",accuracy_score(y_pred=y_pred,y_true = y_test))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.metrics import make_scorer
from sklearn.tree import DecisionTreeClassifier

def fit_model_k_fold(X, y):
    k_fold = KFold(n_splits=10)
    clf = LogisticRegression(random_state=0)
    
    params = {'C':range(1,5,1),'max_iter':range(1,15,1),
             'tol':[0.000001,0.00001,0.0001]}
    
    scoring_fnc = make_scorer(accuracy_score)
    grid = GridSearchCV(clf, param_grid=params,scoring=scoring_fnc,cv=k_fold)
    grid = grid.fit(X, y)
    return grid.best_estimator_

result = fit_model_k_fold(X_test,y_test)
print(result)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_t_pred = model.predict(X_train)

print("训练集准确率为：\n",accuracy_score(y_true=y_train,y_pred=y_t_pred))
print("测试集准确率为：\n",accuracy_score(y_true=y_test,y_pred=y_pred))
```

```python
def fit_model_k_fold(X, y):
    clf = SVC(kernel="rbf",random_state=0)
    
    params = {'gamma':[0.1,0.12,0.14,0.16,0.18,0.2,0.22,0.24,0.26,0.28,0.3],
              'C':[1.1,1.12,1.14,1.16,1.18,1.2,1.22,1.24,1.26,1.28,1.3,1.32,1.34,1.36,1.38,1.4]}
    
    scoring_fnc = make_scorer(accuracy_score)
    grid = GridSearchCV(clf, param_grid=params,scoring=scoring_fnc)
    grid = grid.fit(X, y)
    return grid.best_estimator_
```

```python
result = fit_model_k_fold(X_train,y_train)
print(result)
```

```python
model = SVC(C=1.1, degree=3, gamma=0.22, kernel='rbf')

model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_t_pred = model.predict(X_train)

print("训练集准确率为：\n",accuracy_score(y_true=y_train,y_pred=y_t_pred))
print("测试集准确率为：\n",accuracy_score(y_true=y_test,y_pred=y_pred))
```
